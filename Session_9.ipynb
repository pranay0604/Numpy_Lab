Linear Algebra
--------------

- Dot product, matrix multiplication
        
- Determinant, inverse, eigenvalues/eigenvectors (np.linalg)

- Solving linear equations (np.linalg.solve)

```python
import numpy as np

arr=np.array([7,8,9]) 

arr1=np.array([1,2,3]) 

arr2=np.array([[7,5,6],
               [1,2,3]]) 

np.multiply(arr,arr1)
```

Matrix with Different Shapes
----------------------------
vector <-> vector => scaler

## Note: The dot product always use in between two vectors and return scaler values

vector <-> matrix  => vector(scaler) / 1D array

matrix <-> matrix => matrix

## Note : it should contain column of the first matrix equal to row  of the second matrix.

```python
np.dot(arr,arr1)
```

```python
arr@arr2
```

```python
np.matmul(arr,arr2)
```

```python
x=np.arange(12).reshape(3,4)
y=np.arange(8).reshape(4,2)
```

```python
np.dot(x,y)
```

```python
np.matmul(x,y)
```

```python
a=np.array([7,8,9])
b=np.array([[4],
            [8],
            [6]])

np.matmul(a,b)
```

```python
np.dot(a,b)
```

```python
a@b
```

```python
x=np.array([[7,8,9],
            [1,2,3],
            [4,6,8]])
```

```python
np.trace(x) # Sum of diagonal values
```

```python
np.linalg.matrix_rank(x)
```

```python
np.linalg.det(x)
```

```python
np.linalg.inv(x)

# if the matrix det==0 then the inverse not applicable.
```

```python
x=np.array([[7,8,9],
            [1,2,3],
            [4,10,8]])

np.trace(x)
```

```python
np.linalg.det(x)
```

```python
np.linalg.inv(x)
```

```python
x=np.array([[7,8,9],
            [1,2,3],
            ])

np.trace(x)
```

```python
np.linalg.det(x) 
# The dot value always applicable on square matrix.
```

```python
x=np.array([[2,1],
   [3,2]])
y=np.array([12,15])

np.linalg.solve(x,y)
```

```python
np.linalg.eigvals(x)
```

```python
np.linalg.eig(x)
```
