**Difference between Index and Multi (Fancy) Index**

**Index** :
- It returns only one position
- Index always return specific position value
- By using of index we can update the value

**Multi Index** : 
- It returns number of positions
- It returns specific number of positions
- By using of multi index we can update multiple values

```python
arr = np.arange(16).reshape(4,4)
arr
```

```python
arr[2]
```

```python
arr[0][2]
```

```python
arr[0,2]
```

```python
l = [1,2,3,4,5]
l[0] = 50
l
```

```python
arr[[0,2],[3,1]]=20,30
arr
```

```python
arr[0,2:4]
```

```python
arr[0,2:4]=100
arr
```

```python
arr[[2,3],[2,3]] # points (2,2) & (3,3)
```

```python
arr[1:3,2:4]
```

```python
arr[1:3,2:4]=150
arr
```

```python
arr[1:3,2:4]=150,250
arr
```

**Boolean Index**

- index : it returns specific values with those positions

- boolean : return values with those booleans

```python
arr = np.array([1,2,3,4,5])

arr[[True,False,True,False,True]]
```

```python
res = arr%2 == 0
print(arr[res])
```

```python
arr%2 == 0
```

```python
arr[arr%2==0] #arr[mask] --> return values
```
