How to create array with null values
-----------

```python
arr = np.array([[7,np.nan,8,np.inf],[10,np.inf,4,9]])
arr
```

To return boolean index / mask
--------

```python
# np.isnan(arr)
# np.isposinf(arr)
# np.isneginf(arr)
```

```python
print("NaN values :",np.isnan(arr))
print()
print("Positive inf. :",np.isposinf(arr))
print()
print("Negative inf. :",np.isneginf(arr))
print()
```

To return True values
------------

```python
print("NaN values :",arr[np.isnan(arr)])
print()
print("Positive inf. :",arr[np.isposinf(arr)])
print()
print("Negative inf. :",arr[np.isneginf(arr)])
print()
```

Cleaning / Removing the Null values
----

```python
arr1 = arr[~np.isnan(arr)]
arr1
```

```python
arr2 = arr1[~np.isinf(arr1)]
arr2
```
