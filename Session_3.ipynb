### What is the difference between reshape, ravel, flatten

- **reshape** => Returns an array containing the same data with a new shape, but this is the view

- **ravel** => Return a flattened array and this is also view

- **flatten** => Return a copy of the array collapsed into one dimension.

```python
arr1=np.array([[7,4,6],[1,2,3]])
```

```python
arr1
```

```python
# reshape

res=arr1.reshape(-1)
res[0]=100
print(res)
print(arr1)
```

```python
# ravel 

rav=arr1.ravel()
rav[1]=120
print(arr1)
print(rav)
```

```python
# flatten

flat=arr1.flatten()
flat[1]=50
print(arr1)
print(flat)
```

```python
temps = np.array([32, 35, 31, 30, 29, 28, 33])
temps
```

```python
# 1. Get the temperature on Day 1 and Day 7.
temps[[0,6]]
```

```python
# 2. Get the last 3 days’ temperatures
temps[-3:]
```

```python
# 3. Get Day 2 to Day 5 temperatures.
    
temps[1:-2]
```

```python
# 4. Reverse the week’s data.
    
temps[::-1]
```

```python
# 5. Get every alternate day’s temperature.

temps[::2]
```

```python
temps[2:-2:-1]
```

```python
[i for i in range(1,10)] # 0,...10

# step =1 , start<stop
           # 0<10
```

```python
[i for i in range(10,1)] # 0,...10

# step =1 , start<stop
           # 10<1 # wrong
```

```python
[i for i in range(10,1,-1)] # 0,...10

# step =-1 , start>stop
           # 10>1 # CORRECT
```

```python
[i for i in range(1,10,-1)] # 0,...10

# step =-1 , start>stop
           # 1>10 # wrong
```
