```python
import numpy as np
```

```python
arr = np.arange(20).reshape(4,5)
arr
```

```python
arr%2==0 # mask or boolean values with conditions # array supports vectorization (element wise operation)
```

```python
arr[arr%2==0]
```

```python
arr[arr%5==0]
```

```python
arr[(arr%2==0) & (arr%5==0)]
```

```python
arr[(arr%2==0) | (arr%5==0)]
```

```python
arr[(arr%2==0) ^ (arr%5==0)]
```

```python
arr[(arr%2==0) & (arr%5==0)] = 100
arr
```

```python
arr = np.arange(20).reshape(4,5)
arr
```

```python
np.where(arr%2==0) # [(rows) (columns)]
```

```python
x,y = np.where(arr%2==0) # [(rows) (columns)]
print(x)
print(y)
```

```python
list(zip(x.tolist(),y.tolist()))
```

```python
np.where(arr%2==0,True,False)
```

```python
np.where(arr%2==0,'Even','Odd')
```
