### The difference between the list and numpy performance

```python
import time
import sys
import numpy as np
```

```python
start_time = time.time()
l = [i*2 for i in range(100)]
end_time = time.time()
print("The list time is {:f}".format(end_time-start_time))
print('The size of list :',sys.getsizeof(l))
```

```python
start_time = time.time()
l = np.array([i*2 for i in range(100)])
end_time = time.time()
print("The list time is {:f}".format(end_time-start_time))
print('The size of list :',sys.getsizeof(np))
```
