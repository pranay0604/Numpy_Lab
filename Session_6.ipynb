**Broadcasting**

- the process of arithmetic operation in different shape

- Rules :

- the both arays are compatible :-

    - if the both arrays are same shape

    - Any one has one value

- (7,3)+(7,3) # True

- (7,3)+(1,1) # True

- (3,2)+(2,3) # False


```python
import numpy as np

a = np.array([[4,5,6],[1,2,3]])
b = np.array([[4,5,6],[1,2,3]])
a.shape, b.shape
```

```python
a+b
```

```python
a = np.array([[4,5,6],[1,2,3]])
b = np.array([[4,5],[1,2],[7,8]])
a.shape, b.shape
```

```python
a+b
```

```python
a = np.array([[4,5,6],[1,2,3]])
b = np.array([[4]])
a.shape, b.shape
```

```python
a = np.array([[4,5,6],[1,2,3]])
b = np.array([[1,2,3]])
a.shape, b.shape
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(10,22).reshape(3,4)

a+b
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(10,22).reshape(4,3)

a+b
```

**Note** : a=(4,3), b=(3,4), When you compare coresponding values from left to right (x1)4 != (x2)3 and (y1)3 != (y2)4

**resize**

- change the size of the array

- Return a new array with the specified shape.

- reshape does not take any duplicates

- resize takes duplicates when size is much longer than the array

```python
a = np.array([1,2,3,4,5])
a
```

```python
np.resize(a,(2,4))
```
